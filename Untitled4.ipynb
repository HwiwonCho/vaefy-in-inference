```python
# ============================================================
# STEP 1 (Stage-1): Landscape scan (Qwen)
# - Multi-seed aggregation
# - Robust latent noise injection: last-token only (no prefill/decode shape guess)
# - Metrics:
#   * fmt_valid_mean  = P(Answer format valid)
#   * ans_ok_mean     = P(Answer == 한국)  (fmt 무시)
#   * joint_mean      = P(fmt_valid AND Answer==한국)   <-- MAIN objective
#   * strict_mean     = P(Quote EXACTLY == hidden_fact AND Answer==한국)  (보조)
#   * acc_valid_cond  = P(Answer==한국 | fmt_valid)     (보조)
# - Saves:
#   * CSV (all grid)
#   * best_setting.json (best by joint_mean then fmt then strict then layer proximity tie-break)
#   * heatmaps (joint / fmt / acc|valid / strict)
# ============================================================

import os, re, gc, json
import numpy as np
import pandas as pd
import torch
from tqdm import tqdm
import matplotlib.pyplot as plt
import seaborn as sns
from transformers import AutoModelForCausalLM, AutoTokenizer

# ----------------------------
# 0) Config
# ----------------------------
MODEL_ID = "Qwen/Qwen2.5-1.5B-Instruct"
DTYPE = "auto"
DEVICE = "cuda" if torch.cuda.is_available() else "cpu"

DEPTH = 300
HINT_PCT = 0.95

SIGMA_RANGE = np.arange(0.0, 4.0, 0.4)
LAYER_STRIDE = 2

MAX_NEW_TOKENS = 50

# Multi-seed
N_EVAL_SEEDS = 10
SEED_BASE = 1000
EVAL_SEEDS = [SEED_BASE + i for i in range(N_EVAL_SEEDS)]

SAVE_DIR = "stage1_landscape_qwen"
os.makedirs(SAVE_DIR, exist_ok=True)

OUT_CSV = os.path.join(SAVE_DIR, "noise_landscape_multiseed_joint.csv")
OUT_BEST_JSON = os.path.join(SAVE_DIR, "best_setting.json")

OUT_PNG_JOINT  = os.path.join(SAVE_DIR, "heatmap_joint_mean.png")
OUT_PNG_FMT    = os.path.join(SAVE_DIR, "heatmap_fmt_valid_mean.png")
OUT_PNG_COND   = os.path.join(SAVE_DIR, "heatmap_acc_valid_cond_mean.png")
OUT_PNG_STRICT = os.path.join(SAVE_DIR, "heatmap_strict_mean.png")

print(f"[Stage-1] MODEL={MODEL_ID} DEPTH={DEPTH} HINT={HINT_PCT} seeds={N_EVAL_SEEDS} device={DEVICE}")

# ----------------------------
# 1) Load model/tokenizer
# ----------------------------
tokenizer = AutoTokenizer.from_pretrained(MODEL_ID)
model = AutoModelForCausalLM.from_pretrained(MODEL_ID, torch_dtype=DTYPE, device_map="auto")
model.eval()

if tokenizer.pad_token is None:
    tokenizer.pad_token = tokenizer.eos_token
tokenizer.padding_side = "left"

def get_layers(m):
    if hasattr(m, "model") and hasattr(m.model, "layers"):
        return m.model.layers
    if hasattr(m, "transformer") and hasattr(m.transformer, "h"):
        return m.transformer.h
    raise ValueError("Cannot locate transformer layers.")

layers = get_layers(model)
NUM_LAYERS = len(layers)
LAYER_RANGE = list(range(0, NUM_LAYERS, LAYER_STRIDE))
print(f"[Stage-1] Detected layers={NUM_LAYERS} | scanning layers={LAYER_RANGE}")

# ----------------------------
# 2) Build prompt (Chat template) - fixed
# ----------------------------
hidden_fact = "chris의 국적은 한국인입니다."
bias_context = [f"사람{i}의 국적은 미국인입니다." for i in range(DEPTH)]
hint_pos = int(DEPTH * HINT_PCT)

full_context_list = bias_context[:hint_pos] + [hidden_fact] + bias_context[hint_pos:]
context_text = "\n".join(full_context_list)

user_query = (
    f"아래 문맥을 읽고 질문에 답하시오.\n\n"
    f"--- 문맥 시작 ---\n{context_text}\n--- 문맥 끝 ---\n\n"
    f"질문: chris의 국적은 무엇인가?\n"
    f"규칙:\n"
    f"1) 문맥에서 chris에 관한 문장을 'Quote: ...' 형식으로 정확히 인용하라.\n"
    f"2) 그 다음 줄에 국적을 'Answer: 한국' 또는 'Answer: 미국' 형식으로 답하라.\n"
    f"3) 다른 말은 하지 마라."
)

messages = [
    {"role": "system", "content": "You are a helpful assistant who follows instructions strictly."},
    {"role": "user", "content": user_query},
]

prompt = tokenizer.apply_chat_template(messages, tokenize=False, add_generation_prompt=True)
inputs = tokenizer([prompt], return_tensors="pt", padding=True).to(model.device)

# ----------------------------
# 3) Parsers / scoring
# ----------------------------
ANS_ONLY_RE = re.compile(r"Answer\s*:\s*(한국|미국)")
QUOTE_RE = re.compile(r"Quote\s*:\s*(.+)", re.IGNORECASE)

def _norm_space(s: str) -> str:
    return re.sub(r"\s+", " ", s).strip()

def eval_one(decoded_text: str):
    t = decoded_text.strip()

    # Answer
    m = ANS_ONLY_RE.search(t)
    fmt_valid = (m is not None)
    parsed = m.group(1) if m else None

    ans_ok = bool(parsed == "한국")         # fmt 무시한 "정답 여부"
    joint_ok = bool(fmt_valid and ans_ok)  # MAIN objective

    # Strict: Quote가 hidden_fact와 정확히 동일 + ans_ok
    qm = QUOTE_RE.search(t)
    quote = qm.group(1).strip() if qm else None
    quote_ok = False
    if quote is not None:
        quote_ok = (_norm_space(quote) == _norm_space(hidden_fact))
    strict_ok = bool(quote_ok and ans_ok)

    return {
        "fmt_valid": int(fmt_valid),
        "ans_ok": int(ans_ok),
        "joint_ok": int(joint_ok),
        "strict_ok": int(strict_ok),
    }

# ----------------------------
# 4) Noise Controller (ROBUST): last-token only
# ----------------------------
class NoiseController:
    def __init__(self, model, layer_idx, noise_std=0.0, seed=0):
        self.model = model
        self.layer_idx = int(layer_idx)
        self.noise_std = float(noise_std)
        self.seed = int(seed)
        self.hook_handle = None
        self.gen = None

    def _hook(self, module, inputs, output):
        if self.noise_std <= 0:
            return output

        hs = output[0] if isinstance(output, tuple) else output

        if self.gen is None:
            self.gen = torch.Generator(device=hs.device).manual_seed(self.seed)

        # (B,T,H) -> perturb last token only; (B,H) -> perturb whole
        if hs.dim() == 3:
            last = hs[:, -1:, :]
            noise = torch.randn(last.size(), generator=self.gen, device=hs.device, dtype=hs.dtype) * self.noise_std
            hs2 = torch.cat([hs[:, :-1, :], last + noise], dim=1)
        elif hs.dim() == 2:
            noise = torch.randn(hs.size(), generator=self.gen, device=hs.device, dtype=hs.dtype) * self.noise_std
            hs2 = hs + noise
        else:
            return output

        return (hs2,) + output[1:] if isinstance(output, tuple) else hs2

    def attach(self):
        self.hook_handle = get_layers(self.model)[self.layer_idx].register_forward_hook(self._hook)

    def detach(self):
        if self.hook_handle:
            self.hook_handle.remove()
            self.hook_handle = None
        self.gen = None

# ----------------------------
# 5) Generate wrapper (greedy)
# ----------------------------
@torch.inference_mode()
def generate_one(trial_seed: int):
    torch.manual_seed(trial_seed)
    if torch.cuda.is_available():
        torch.cuda.manual_seed_all(trial_seed)

    out = model.generate(
        **inputs,
        max_new_tokens=MAX_NEW_TOKENS,
        do_sample=False,
        pad_token_id=tokenizer.pad_token_id,
        eos_token_id=tokenizer.eos_token_id,
    )

    in_len = inputs.input_ids.shape[1]
    gen_part = out[:, in_len:]
    return tokenizer.decode(gen_part[0], skip_special_tokens=True)

# ----------------------------
# 6) Scan (layer × sigma) with MULTI-SEED aggregation
# ----------------------------
rows = []
pbar = tqdm(total=len(LAYER_RANGE) * len(SIGMA_RANGE), dynamic_ncols=True, desc="Stage-1 scan (multi-seed)")

for layer_idx in LAYER_RANGE:
    for sigma in SIGMA_RANGE:
        fmt_list = []
        ans_list = []
        joint_list = []
        strict_list = []

        for s in EVAL_SEEDS:
            ctrl = NoiseController(model, layer_idx, noise_std=sigma, seed=s)
            ctrl.attach()
            try:
                text = generate_one(s)
            finally:
                ctrl.detach()

            met = eval_one(text)
            fmt_list.append(met["fmt_valid"])
            ans_list.append(met["ans_ok"])
            joint_list.append(met["joint_ok"])
            strict_list.append(met["strict_ok"])

        fmt_valid_mean = float(np.mean(fmt_list))
        ans_ok_mean = float(np.mean(ans_list))
        joint_mean = float(np.mean(joint_list))      # MAIN
        strict_mean = float(np.mean(strict_list))    # AUX

        valid_cnt = int(np.sum(fmt_list))
        if valid_cnt > 0:
            acc_valid_cond_mean = float(np.sum([a for a, f in zip(ans_list, fmt_list) if f == 1]) / valid_cnt)
        else:
            acc_valid_cond_mean = np.nan

        rows.append({
            "layer": layer_idx,
            "sigma": float(sigma),
            "n_seeds": N_EVAL_SEEDS,
            "valid_cnt": valid_cnt,
            "fmt_valid_mean": fmt_valid_mean,
            "ans_ok_mean": ans_ok_mean,
            "joint_mean": joint_mean,
            "acc_valid_cond_mean": acc_valid_cond_mean,
            "strict_mean": strict_mean,
        })

        pbar.set_postfix(L=layer_idx, s=f"{sigma:.1f}", joint=f"{joint_mean:.2f}", fmt=f"{fmt_valid_mean:.2f}")
        pbar.update(1)

pbar.close()

df = pd.DataFrame(rows)
df.to_csv(OUT_CSV, index=False)
print(f"[Stage-1] Saved: {OUT_CSV}")

# ----------------------------
# 7) Pick BEST by JOINT MEAN
#   - tie-break: fmt_valid_mean, strict_mean, then prefer smaller sigma, then middle-ish layer (optional stability)
# ----------------------------
df_sorted = df.sort_values(
    by=["joint_mean", "fmt_valid_mean", "strict_mean", "sigma"],
    ascending=[False, False, False, True],
).reset_index(drop=True)

best = df_sorted.head(1)
best_layer = int(best.iloc[0]["layer"])
best_sigma = float(best.iloc[0]["sigma"])

best_payload = {
    "model": MODEL_ID,
    "depth": DEPTH,
    "hint_pct": HINT_PCT,
    "metric": "joint_mean = P(FormatValid AND Answer==한국)",
    "n_eval_seeds": N_EVAL_SEEDS,
    "seed_base": SEED_BASE,
    "best_layer": best_layer,
    "best_sigma": best_sigma,
    "best_joint_mean": float(best.iloc[0]["joint_mean"]),
    "best_fmt_valid_mean": float(best.iloc[0]["fmt_valid_mean"]),
    "best_strict_mean": float(best.iloc[0]["strict_mean"]),
    "best_row": best.iloc[0].to_dict(),
}

with open(OUT_BEST_JSON, "w", encoding="utf-8") as f:
    json.dump(best_payload, f, ensure_ascii=False, indent=2)

print("\n=== BEST (Stage-1) ===")
print(best_payload)
print(f"[Stage-1] Saved: {OUT_BEST_JSON}")

# ----------------------------
# 8) Heatmaps
# ----------------------------
def save_heatmap(value_col: str, out_png: str, title: str):
    piv = df.pivot(index="layer", columns="sigma", values=value_col)
    plt.figure(figsize=(12, 6))
    sns.heatmap(piv, cmap="viridis", annot=True, fmt=".2f")
    plt.title(title)
    plt.ylabel("Layer")
    plt.xlabel("Noise Sigma")
    plt.tight_layout()
    plt.savefig(out_png, dpi=180)
    plt.show()
    print(f"[Stage-1] Saved: {out_png}")

save_heatmap("joint_mean", OUT_PNG_JOINT,  f"joint_mean = P(FormatValid & Correct) | DEPTH={DEPTH} HINT={HINT_PCT}")
save_heatmap("fmt_valid_mean", OUT_PNG_FMT, f"fmt_valid_mean | DEPTH={DEPTH} HINT={HINT_PCT}")
save_heatmap("acc_valid_cond_mean", OUT_PNG_COND, f"acc_valid_cond_mean = P(Correct | FormatValid) | DEPTH={DEPTH} HINT={HINT_PCT}")
save_heatmap("strict_mean", OUT_PNG_STRICT, f"strict_mean (Quote==Fact & Correct) | DEPTH={DEPTH} HINT={HINT_PCT}")

```

```python
# ============================================================
# STEP 2 (Stage-2): Robustness sweep (Qwen) — CLEAN FINAL
# - Uses best_layer/best_sigma from Stage-1 JSON
# - Prompt matched with Stage-1 (Quote + Answer), BUT we evaluate only:
#     * Valid (Answer format)
#     * Correct (Answer==한국 AND Valid)
#     * Joint = Valid AND Correct   <-- MAIN metric (Format & Accurate)
#     * Correct|Valid (analysis)
# - Heatmaps (gasi-seong improved):
#   For each metric, draw 1x3 panels side-by-side:
#     [Baseline] [Temp] [Latent Noise]
# ============================================================

import os, re, json
import numpy as np
import pandas as pd
import torch
from tqdm import tqdm
import matplotlib.pyplot as plt
import seaborn as sns
from transformers import AutoModelForCausalLM, AutoTokenizer

# ----------------------------
# 0) Paths
# ----------------------------
STAGE1_BEST_JSON = "stage1_landscape_qwen/best_setting.json"
SAVE_DIR = "stage2_robustness_qwen"
os.makedirs(SAVE_DIR, exist_ok=True)

OUT_RUNS_CSV = os.path.join(SAVE_DIR, "runs_matched.csv")
OUT_SUMM_CSV = os.path.join(SAVE_DIR, "summary_matched.csv")

OUT_FIG_VALID = os.path.join(SAVE_DIR, "hm_valid_methods.png")
OUT_FIG_CORR  = os.path.join(SAVE_DIR, "hm_correct_methods.png")
OUT_FIG_JOINT = os.path.join(SAVE_DIR, "hm_joint_methods.png")          # ★ MAIN
OUT_FIG_COND  = os.path.join(SAVE_DIR, "hm_correct_given_valid.png")    # analysis

# ----------------------------
# 1) Sweep config
# ----------------------------
MODEL_ID = "Qwen/Qwen2.5-1.5B-Instruct"
DTYPE = "auto"
DEVICE = "cuda" if torch.cuda.is_available() else "cpu"

DEPTHS = [50, 100, 300]
HINT_PCTS = [0.10, 0.50, 0.95]
N_SEEDS = 30
SEED_BASE = 1000
MAX_NEW_TOKENS = 50

BEST_TEMP = 1.2
TOP_P = 0.95

print(f"[Stage-2] MODEL={MODEL_ID} device={DEVICE} depths={DEPTHS} hints={HINT_PCTS} seeds={N_SEEDS}")

# ----------------------------
# 2) Load best (layer/sigma)
# ----------------------------
if not os.path.exists(STAGE1_BEST_JSON):
    raise FileNotFoundError(f"Stage-1 best json not found: {STAGE1_BEST_JSON}")

with open(STAGE1_BEST_JSON, "r", encoding="utf-8") as f:
    best_config = json.load(f)

TARGET_LAYER = int(best_config["best_layer"])
BEST_SIGMA = float(best_config["best_sigma"])
print(f"[Stage-2] Loaded Best: Layer={TARGET_LAYER}, Sigma={BEST_SIGMA}")

# ----------------------------
# 3) Load model/tokenizer
# ----------------------------
tokenizer = AutoTokenizer.from_pretrained(MODEL_ID)
model = AutoModelForCausalLM.from_pretrained(MODEL_ID, device_map="auto", torch_dtype=DTYPE)
model.eval()

if tokenizer.pad_token is None:
    tokenizer.pad_token = tokenizer.eos_token
tokenizer.padding_side = "left"

def get_layers(m):
    if hasattr(m, "model") and hasattr(m.model, "layers"):
        return m.model.layers
    if hasattr(m, "transformer") and hasattr(m.transformer, "h"):
        return m.transformer.h
    raise ValueError("Cannot locate transformer layers.")

# ----------------------------
# 4) Robust NoiseController (last-token only)
# ----------------------------
class NoiseController:
    def __init__(self, model, layer_idx, sigma, seed):
        self.model = model
        self.layer_idx = int(layer_idx)
        self.sigma = float(sigma)
        self.seed = int(seed)
        self.hook = None
        self.gen = None

    def _hook_fn(self, module, inp, out):
        if self.sigma <= 0:
            return out

        hs = out[0] if isinstance(out, tuple) else out

        if self.gen is None:
            self.gen = torch.Generator(device=hs.device).manual_seed(self.seed)

        if hs.dim() == 3:
            last = hs[:, -1:, :]
            noise = torch.randn(last.size(), generator=self.gen, device=hs.device, dtype=hs.dtype) * self.sigma
            hs2 = torch.cat([hs[:, :-1, :], last + noise], dim=1)
        elif hs.dim() == 2:
            noise = torch.randn(hs.size(), generator=self.gen, device=hs.device, dtype=hs.dtype) * self.sigma
            hs2 = hs + noise
        else:
            return out

        return (hs2,) + out[1:] if isinstance(out, tuple) else hs2

    def __enter__(self):
        self.hook = get_layers(self.model)[self.layer_idx].register_forward_hook(self._hook_fn)
        return self

    def __exit__(self, exc_type, exc_val, exc_tb):
        if self.hook:
            self.hook.remove()
        self.gen = None

# ----------------------------
# 5) Prompt builder (MATCHED with Stage-1)
# ----------------------------
HIDDEN_FACT = "chris의 국적은 한국인입니다."

def build_prompt(depth: int, hint_pct: float):
    bias_context = [f"사람{i}의 국적은 미국인입니다." for i in range(depth)]
    hint_pos = int(depth * hint_pct)
    full_context = bias_context[:hint_pos] + [HIDDEN_FACT] + bias_context[hint_pos:]
    context_text = "\n".join(full_context)

    user_query = (
        f"아래 문맥을 읽고 질문에 답하시오.\n\n"
        f"--- 문맥 시작 ---\n{context_text}\n--- 문맥 끝 ---\n\n"
        f"질문: chris의 국적은 무엇인가?\n"
        f"규칙:\n"
        f"1) 문맥에서 chris에 관한 문장을 'Quote: ...' 형식으로 정확히 인용하라.\n"
        f"2) 그 다음 줄에 국적을 'Answer: 한국' 또는 'Answer: 미국' 형식으로 답하라.\n"
        f"3) 다른 말은 하지 마라."
    )

    messages = [
        {"role": "system", "content": "You are a helpful assistant who follows instructions strictly."},
        {"role": "user", "content": user_query},
    ]
    return tokenizer.apply_chat_template(messages, tokenize=False, add_generation_prompt=True)

# ----------------------------
# 6) Parsers (QUOTE/STRICT ignored)
# ----------------------------
ANS_ONLY_RE = re.compile(r"Answer\s*:\s*(한국|미국)")

def parse_valid_correct(text: str):
    """
    valid  := Answer: (한국|미국) is present
    correct:= valid AND Answer==한국
    joint  := valid AND correct (== correct here, but kept explicit for clarity)
    """
    t = text.strip()
    m = ANS_ONLY_RE.search(t)
    valid = int(m is not None)
    parsed = m.group(1) if m else None
    correct = int(valid and (parsed == "한국"))
    joint = int(valid and correct)
    return valid, correct, joint

# ----------------------------
# 7) Generation wrapper
# ----------------------------
@torch.inference_mode()
def run_generate(prompt, use_noise, use_sample, temp, seed):
    torch.manual_seed(seed)
    if torch.cuda.is_available():
        torch.cuda.manual_seed_all(seed)

    inputs = tokenizer([prompt], return_tensors="pt", padding=True).to(model.device)
    in_len = inputs.input_ids.shape[1]

    gen_kwargs = dict(
        max_new_tokens=MAX_NEW_TOKENS,
        pad_token_id=tokenizer.pad_token_id,
        eos_token_id=tokenizer.eos_token_id,
    )

    if use_sample:
        gen_kwargs.update(dict(do_sample=True, temperature=temp, top_p=TOP_P))
    else:
        gen_kwargs.update(dict(do_sample=False))

    if use_noise:
        with NoiseController(model, TARGET_LAYER, BEST_SIGMA, seed):
            out = model.generate(**inputs, **gen_kwargs)
    else:
        out = model.generate(**inputs, **gen_kwargs)

    return tokenizer.decode(out[0][in_len:], skip_special_tokens=True)

# ----------------------------
# 8) Experiment loop
# ----------------------------
experiment_modes = [
    ("Baseline (Greedy)", dict(use_noise=False, use_sample=False, temp=0.0)),
    (f"Temp Sampling (T={BEST_TEMP})", dict(use_noise=False, use_sample=True, temp=BEST_TEMP)),
    (f"Latent Noise (σ={BEST_SIGMA})", dict(use_noise=True, use_sample=False, temp=0.0)),
]

rows = []
total = len(DEPTHS) * len(HINT_PCTS) * len(experiment_modes) * N_SEEDS
pbar = tqdm(total=total, desc="Stage-2 sweep", dynamic_ncols=True)

for depth in DEPTHS:
    for hint_pct in HINT_PCTS:
        prompt = build_prompt(depth, hint_pct)

        for method, cfg in experiment_modes:
            for i in range(N_SEEDS):
                seed = SEED_BASE + i
                out_text = run_generate(prompt, cfg["use_noise"], cfg["use_sample"], cfg["temp"], seed)

                valid, correct, joint = parse_valid_correct(out_text)

                rows.append({
                    "Depth": int(depth),
                    "HintPct": float(hint_pct),
                    "Method": method,
                    "Seed": int(seed),
                    "Valid": int(valid),
                    "Correct": int(correct),
                    "Joint": int(joint),      # ★ MAIN (Format & Accurate)
                    "Output": out_text,
                })
                pbar.update(1)

pbar.close()

df = pd.DataFrame(rows)
df.to_csv(OUT_RUNS_CSV, index=False)
print(f"[Stage-2] Saved: {OUT_RUNS_CSV}")

# ----------------------------
# 9) Summary (mean over seeds) + conditional
# ----------------------------
summ = (
    df.groupby(["Depth", "HintPct", "Method"])[["Valid", "Correct", "Joint"]]
      .mean()
      .reset_index()
)

summ["ValidPct"] = summ["Valid"] * 100
summ["CorrectPct"] = summ["Correct"] * 100
summ["JointPct"] = summ["Joint"] * 100

def cond_acc(subdf):
    v = int(subdf["Valid"].sum())
    if v == 0:
        return np.nan
    return float(subdf["Correct"].sum() / v)

cond = (
    df.groupby(["Depth", "HintPct", "Method"])
      .apply(cond_acc)
      .reset_index(name="CorrectGivenValid")
)
cond["CorrectGivenValidPct"] = cond["CorrectGivenValid"] * 100

summ2 = summ.merge(cond, on=["Depth", "HintPct", "Method"], how="left")
summ2.to_csv(OUT_SUMM_CSV, index=False)
print(f"[Stage-2] Saved: {OUT_SUMM_CSV}")

print("\n=== Summary (mean over seeds) ===")
print(summ2[["Depth","HintPct","Method","ValidPct","CorrectPct","JointPct","CorrectGivenValidPct"]])

# ----------------------------
# 10) Heatmaps: methods side-by-side (1x3 panels)
#   - Valid (%)
#   - Correct (%)
#   - Joint (%): Format & Accurate  ★ MAIN
#   - Correct|Valid (%): analysis
# ----------------------------
def _pivot_mean(df_in: pd.DataFrame, value_col: str):
    return (
        df_in.groupby(["Depth", "HintPct"])[value_col]
            .mean()
            .reset_index()
            .pivot(index="Depth", columns="HintPct", values=value_col)
            .sort_index()
    )

def _pivot_cond(df_in: pd.DataFrame):
    # Correct|Valid per (Depth, HintPct)
    rows = []
    for (d, h), g in df_in.groupby(["Depth", "HintPct"]):
        v = int(g["Valid"].sum())
        c = int(g["Correct"].sum())
        rows.append({"Depth": d, "HintPct": h, "Cond": (c / v) if v > 0 else np.nan})
    piv = pd.DataFrame(rows).pivot(index="Depth", columns="HintPct", values="Cond").sort_index()
    return piv

def plot_methods_row(metric_name: str, pivots: dict, out_png: str, vmin=None, vmax=None):
    """
    pivots: dict(method -> pivot_table with index Depth, columns HintPct, values in [0,1] or nan)
    """
    methods = list(pivots.keys())

    # global color scale (recommended for fairness)
    if vmin is None:
        vmin = np.nanmin([p.values for p in pivots.values()])
    if vmax is None:
        vmax = np.nanmax([p.values for p in pivots.values()])

    fig, axes = plt.subplots(1, len(methods), figsize=(4.8 * len(methods), 4.2), constrained_layout=True)

    if len(methods) == 1:
        axes = [axes]

    for ax, method in zip(axes, methods):
        piv = pivots[method]
        sns.heatmap(
            piv * 100,
            ax=ax,
            annot=True,
            fmt=".1f",
            cmap="viridis",
            vmin=vmin * 100,
            vmax=vmax * 100,
            cbar=(ax == axes[-1]),  # colorbar only on last panel
        )
        ax.set_title(method)
        ax.set_xlabel("HintPct")
        ax.set_ylabel("Depth")

    fig.suptitle(metric_name, y=1.02)
    plt.savefig(out_png, dpi=200, bbox_inches="tight")
    plt.show()
    print(f"[Stage-2] Saved: {out_png}")

# Build pivot tables per method
methods = [m for (m, _) in experiment_modes]

piv_valid = {m: _pivot_mean(df[df["Method"] == m], "Valid") for m in methods}
piv_corr  = {m: _pivot_mean(df[df["Method"] == m], "Correct") for m in methods}
piv_joint = {m: _pivot_mean(df[df["Method"] == m], "Joint") for m in methods}
piv_cond  = {m: _pivot_cond(df[df["Method"] == m]) for m in methods}

# Plot: side-by-side
plot_methods_row("Valid Rate (%)", piv_valid, OUT_FIG_VALID, vmin=0.0, vmax=1.0)
plot_methods_row("Correct Rate (%)", piv_corr, OUT_FIG_CORR, vmin=0.0, vmax=1.0)
plot_methods_row("Joint Rate (%) = P(FormatValid & Correct)", piv_joint, OUT_FIG_JOINT, vmin=0.0, vmax=1.0)
plot_methods_row("Correct | Valid (%)", piv_cond, OUT_FIG_COND, vmin=0.0, vmax=1.0)

```

```python
# ============================================================
# Stage-1 (LLAMA): Landscape scan (MULTI-DEPTH) - FAST FINAL
# - Prefill cached ONCE per depth
# - FIX: Prefill uses inputs[:-1], Decode starts with inputs[-1]
# - Decode-only generation for each (layer, sigma, seed)
# - Robust latent noise injection: last-token only (decode-step only)
# - Removes padding (batch=1) to avoid position/mask edge cases
# - Metrics:
#   * fmt_valid_mean      = P(Answer format valid)
#   * ans_ok_mean         = P(Answer == 한국)  (fmt 무시)
#   * joint_mean          = P(fmt_valid AND Answer==한국)   <-- MAIN
#   * strict_mean         = P(Quote EXACTLY == hidden_fact AND Answer==한국)
#   * acc_valid_cond_mean = P(Answer==한국 | fmt_valid)
# - Saves per DEPTH:
#   * CSV (grid)
#   * best_setting_depth{DEPTH}.json
#   * heatmaps (joint / fmt / acc|valid / strict)
# ============================================================

import os, re, gc, json
import numpy as np
import pandas as pd
import torch
from tqdm import tqdm
import matplotlib.pyplot as plt
import seaborn as sns
from transformers import AutoModelForCausalLM, AutoTokenizer

# ----------------------------
# 0) Config
# ----------------------------
MODEL_ID = "meta-llama/Meta-Llama-3-8B-Instruct"
DEVICE = "cuda" if torch.cuda.is_available() else "cpu"

DEPTHS = [300]
HINT_PCT = 0.95

# include 0.50 (IMPORTANT)
SIGMA_RANGE = np.arange(0.0, 0.51, 0.01)

LAYER_STRIDE = 2
MAX_NEW_TOKENS = 20

# Multi-seed
N_EVAL_SEEDS = 5
SEED_BASE = 1000
EVAL_SEEDS = [SEED_BASE + i for i in range(N_EVAL_SEEDS)]

TORCH_DTYPE = torch.bfloat16 if torch.cuda.is_available() else torch.float32

SAVE_ROOT = "stage1_landscape_llama_fast_final"
os.makedirs(SAVE_ROOT, exist_ok=True)

print(f"[Stage-1 FAST FINAL] MODEL={MODEL_ID} depths={DEPTHS} HINT={HINT_PCT} seeds={N_EVAL_SEEDS} device={DEVICE}")

# ----------------------------
# 1) Load model/tokenizer
# ----------------------------
tokenizer = AutoTokenizer.from_pretrained(MODEL_ID)
if tokenizer.pad_token is None:
    tokenizer.pad_token = tokenizer.eos_token

model = AutoModelForCausalLM.from_pretrained(
    MODEL_ID,
    torch_dtype=TORCH_DTYPE,
    device_map="auto",
)
model.eval()

def get_layers(m):
    if hasattr(m, "model") and hasattr(m.model, "layers"):
        return m.model.layers
    if hasattr(m, "transformer") and hasattr(m.transformer, "h"):
        return m.transformer.h
    raise ValueError("Cannot locate transformer layers.")

layers = get_layers(model)
NUM_LAYERS = len(layers)
LAYER_RANGE = list(range(0, NUM_LAYERS, LAYER_STRIDE))
print(f"[Stage-1 FAST FINAL] Detected layers={NUM_LAYERS} | scanning layers={LAYER_RANGE}")

# ----------------------------
# 2) Parsers / scoring
# ----------------------------
hidden_fact = "chris의 국적은 한국인입니다."
ANS_ONLY_RE = re.compile(r"Answer\s*:\s*(한국|미국)")
QUOTE_RE = re.compile(r"Quote\s*:\s*(.+)", re.IGNORECASE)

def _norm_space(s: str) -> str:
    return re.sub(r"\s+", " ", s).strip()

def eval_one(decoded_text: str):
    t = decoded_text.strip()

    m = ANS_ONLY_RE.search(t)
    fmt_valid = (m is not None)
    parsed = m.group(1) if m else None

    ans_ok = bool(parsed == "한국")
    joint_ok = bool(fmt_valid and ans_ok)

    qm = QUOTE_RE.search(t)
    quote = qm.group(1).strip() if qm else None
    quote_ok = False
    if quote is not None:
        quote_ok = (_norm_space(quote) == _norm_space(hidden_fact))
    strict_ok = bool(quote_ok and ans_ok)

    return {
        "fmt_valid": int(fmt_valid),
        "ans_ok": int(ans_ok),
        "joint_ok": int(joint_ok),
        "strict_ok": int(strict_ok),
    }

# ----------------------------
# 3) Noise Controller (decode-step last-token only)
# ----------------------------
class NoiseController:
    def __init__(self, model, layer_idx, noise_std=0.0, seed=0):
        self.model = model
        self.layer_idx = int(layer_idx)
        self.noise_std = float(noise_std)
        self.seed = int(seed)
        self.hook_handle = None
        self.gen = None

    def _hook(self, module, inputs, output):
        if self.noise_std <= 0:
            return output

        hs = output[0] if isinstance(output, tuple) else output

        if self.gen is None:
            self.gen = torch.Generator(device=hs.device).manual_seed(self.seed)

        # In decode-only, hs is typically (B,1,H) -> adding noise here == "last token only"
        noise = torch.randn(hs.size(), generator=self.gen, device=hs.device, dtype=hs.dtype) * self.noise_std
        hs2 = hs + noise

        return (hs2,) + output[1:] if isinstance(output, tuple) else hs2

    def attach(self):
        self.hook_handle = get_layers(self.model)[self.layer_idx].register_forward_hook(self._hook)

    def detach(self):
        if self.hook_handle:
            self.hook_handle.remove()
            self.hook_handle = None
        self.gen = None

# ----------------------------
# 4) Prompt builder (NO PADDING for stability)
# ----------------------------
def build_inputs_for_depth(depth: int, hint_pct: float):
    bias_context = [f"사람{i}의 국적은 미국인입니다." for i in range(depth)]
    hint_pos = int(depth * hint_pct)
    full_context_list = bias_context[:hint_pos] + [hidden_fact] + bias_context[hint_pos:]
    context_text = "\n".join(full_context_list)

    user_query = (
        f"아래 문맥을 읽고 질문에 답하시오.\n\n"
        f"--- 문맥 시작 ---\n{context_text}\n--- 문맥 끝 ---\n\n"
        f"질문: chris의 국적은 무엇인가?\n"
        f"규칙:\n"
        f"1) 문맥에서 chris에 관한 문장을 'Quote: ...' 형식으로 정확히 인용하라.\n"
        f"2) 그 다음 줄에 국적을 'Answer: 한국' 또는 'Answer: 미국' 형식으로 답하라.\n"
        f"3) 다른 말은 하지 마라."
    )

    messages = [
        {"role": "system", "content": "You are a helpful assistant who follows instructions strictly."},
        {"role": "user", "content": user_query},
    ]

    prompt = tokenizer.apply_chat_template(messages, tokenize=False, add_generation_prompt=True)

    # IMPORTANT: no padding (batch=1) -> avoids left-pad position_id quirks
    inputs = tokenizer(prompt, return_tensors="pt").to(model.device)
    inputs["attention_mask"] = torch.ones_like(inputs["input_ids"], device=inputs["input_ids"].device)
    return inputs

# ----------------------------
# 5) Prefill ([:-1]) & Decode-only (start from [-1])
# ----------------------------
@torch.inference_mode()
def prefill_cache(input_ids, attention_mask):
    # NO NOISE here (hook not attached during this call)
    out = model(
        input_ids=input_ids,
        attention_mask=attention_mask,
        use_cache=True,
        return_dict=True,
    )
    return out.past_key_values

@torch.inference_mode()
def decode_only_generate(past_key_values, start_token_id, max_new_tokens, trial_seed: int):
    # Determinism setup (though greedy doesn't need randomness; seed is for noise generator)
    torch.manual_seed(trial_seed)
    if torch.cuda.is_available():
        torch.cuda.manual_seed_all(trial_seed)

    generated = []
    cur_input = start_token_id
    curr_past = past_key_values

    for _ in range(max_new_tokens):
        outputs = model(
            input_ids=cur_input,
            use_cache=True,
            past_key_values=curr_past,
            return_dict=True,
        )
        logits = outputs.logits[:, -1, :]
        next_token = torch.argmax(logits, dim=-1, keepdim=True)

        curr_past = outputs.past_key_values
        generated.append(next_token)

        if next_token.item() == tokenizer.eos_token_id:
            break
        cur_input = next_token

    if len(generated) == 0:
        return ""

    gen_ids = torch.cat(generated, dim=1)
    return tokenizer.decode(gen_ids[0], skip_special_tokens=True)

# ----------------------------
# 6) Heatmap helper
# ----------------------------
def save_heatmap(df, value_col: str, out_png: str, title: str):
    piv = df.pivot(index="layer", columns="sigma", values=value_col)
    plt.figure(figsize=(12, 6))
    sns.heatmap(piv, cmap="viridis", annot=False)  # annot=False for speed
    plt.title(title)
    plt.ylabel("Layer")
    plt.xlabel("Noise Sigma")
    plt.tight_layout()
    plt.savefig(out_png, dpi=180)
    plt.close()
    print(f"[Saved] {out_png}")

# ----------------------------
# 7) Main Loop
# ----------------------------
for DEPTH in DEPTHS:
    save_dir = os.path.join(SAVE_ROOT, f"depth{DEPTH}_hint{HINT_PCT}")
    os.makedirs(save_dir, exist_ok=True)

    OUT_CSV = os.path.join(save_dir, "noise_landscape_multiseed_joint.csv")
    OUT_BEST_JSON = os.path.join(save_dir, f"best_setting_depth{DEPTH}.json")

    OUT_PNG_JOINT  = os.path.join(save_dir, "heatmap_joint_mean.png")
    OUT_PNG_FMT    = os.path.join(save_dir, "heatmap_fmt_valid_mean.png")
    OUT_PNG_COND   = os.path.join(save_dir, "heatmap_acc_valid_cond_mean.png")
    OUT_PNG_STRICT = os.path.join(save_dir, "heatmap_strict_mean.png")

    print(f"\n[Stage-1 FAST FINAL] ===== DEPTH={DEPTH} HINT={HINT_PCT} =====")

    full_inputs = build_inputs_for_depth(DEPTH, HINT_PCT)
    full_ids = full_inputs["input_ids"]
    full_mask = full_inputs["attention_mask"]
    full_len = int(full_ids.shape[1])
    print(f"[Depth={DEPTH}] prompt tokens = {full_len}")

    # FIX: Prefill uses prompt[:-1], decode starts with prompt[-1]
    prefill_ids = full_ids[:, :-1]
    prefill_mask = full_mask[:, :-1]
    start_token_id = full_ids[:, -1:]  # last token

    print("[Prefill] caching past_key_values (once, without last token)...")
    past_cache_base = prefill_cache(prefill_ids, prefill_mask)

    rows = []
    total_steps = len(LAYER_RANGE) * len(SIGMA_RANGE)
    pbar = tqdm(total=total_steps, dynamic_ncols=True, desc=f"Scan depth={DEPTH} (decode-only)")

    for layer_idx in LAYER_RANGE:
        # Attach hook once per layer
        ctrl = NoiseController(model, layer_idx, noise_std=0.0, seed=0)
        ctrl.attach()
        try:
            for sigma in SIGMA_RANGE:
                fmt_list, ans_list, joint_list, strict_list = [], [], [], []

                for s in EVAL_SEEDS:
                    # Update hook params
                    ctrl.noise_std = float(sigma)
                    ctrl.seed = int(s)
                    ctrl.gen = None

                    text = decode_only_generate(past_cache_base, start_token_id, MAX_NEW_TOKENS, s)
                    met = eval_one(text)

                    fmt_list.append(met["fmt_valid"])
                    ans_list.append(met["ans_ok"])
                    joint_list.append(met["joint_ok"])
                    strict_list.append(met["strict_ok"])

                fmt_valid_mean = float(np.mean(fmt_list))
                ans_ok_mean = float(np.mean(ans_list))
                joint_mean = float(np.mean(joint_list))
                strict_mean = float(np.mean(strict_list))

                valid_cnt = int(np.sum(fmt_list))
                if valid_cnt > 0:
                    acc_valid_cond_mean = float(np.sum([a for a, f in zip(ans_list, fmt_list) if f == 1]) / valid_cnt)
                else:
                    acc_valid_cond_mean = np.nan

                rows.append({
                    "depth": DEPTH,
                    "hint_pct": HINT_PCT,
                    "layer": int(layer_idx),
                    "sigma": float(sigma),
                    "n_seeds": N_EVAL_SEEDS,
                    "valid_cnt": valid_cnt,
                    "fmt_valid_mean": fmt_valid_mean,
                    "ans_ok_mean": ans_ok_mean,
                    "joint_mean": joint_mean,
                    "acc_valid_cond_mean": acc_valid_cond_mean,
                    "strict_mean": strict_mean,
                })

                pbar.set_postfix(L=layer_idx, s=f"{sigma:.2f}", joint=f"{joint_mean:.2f}", fmt=f"{fmt_valid_mean:.2f}")
                pbar.update(1)

        finally:
            ctrl.detach()

        if torch.cuda.is_available():
            torch.cuda.empty_cache()

    pbar.close()

    df = pd.DataFrame(rows)
    df.to_csv(OUT_CSV, index=False)
    print(f"[Saved] {OUT_CSV}")

    # Pick BEST by joint_mean, then fmt, then strict, then smaller sigma
    df_sorted = df.sort_values(
        by=["joint_mean", "fmt_valid_mean", "strict_mean", "sigma"],
        ascending=[False, False, False, True],
    ).reset_index(drop=True)

    best = df_sorted.head(1)
    best_payload = {
        "model": MODEL_ID,
        "depth": DEPTH,
        "hint_pct": HINT_PCT,
        "metric": "joint_mean = P(FormatValid AND Answer==한국)",
        "n_eval_seeds": N_EVAL_SEEDS,
        "seed_base": SEED_BASE,
        "best_layer": int(best.iloc[0]["layer"]),
        "best_sigma": float(best.iloc[0]["sigma"]),
        "best_joint_mean": float(best.iloc[0]["joint_mean"]),
        "best_fmt_valid_mean": float(best.iloc[0]["fmt_valid_mean"]),
        "best_strict_mean": float(best.iloc[0]["strict_mean"]),
        "best_row": best.iloc[0].to_dict(),
    }

    with open(OUT_BEST_JSON, "w", encoding="utf-8") as f:
        json.dump(best_payload, f, ensure_ascii=False, indent=2)

    print("\n=== BEST (Stage-1 FAST FINAL) ===")
    print(best_payload)
    print(f"[Saved] {OUT_BEST_JSON}")

    # Heatmaps
    save_heatmap(df, "joint_mean", OUT_PNG_JOINT,  f"joint_mean = P(FormatValid & Correct) | DEPTH={DEPTH} HINT={HINT_PCT}")
    save_heatmap(df, "fmt_valid_mean", OUT_PNG_FMT, f"fmt_valid_mean | DEPTH={DEPTH} HINT={HINT_PCT}")
    save_heatmap(df, "acc_valid_cond_mean", OUT_PNG_COND, f"acc_valid_cond_mean = P(Correct | FormatValid) | DEPTH={DEPTH} HINT={HINT_PCT}")
    save_heatmap(df, "strict_mean", OUT_PNG_STRICT, f"strict_mean (Quote==Fact & Correct) | DEPTH={DEPTH} HINT={HINT_PCT}")

    # Cleanup
    del df, rows, df_sorted
    gc.collect()
    if torch.cuda.is_available():
        torch.cuda.empty_cache()

print("\n[Stage-1 FAST FINAL] ALL DONE.")
```
